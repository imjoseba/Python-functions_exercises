# Ejercicio 2
Escriba una función de Python para dar la suma de los números de una lista.

````python
def suma_lista(lista_numeros):
    '''Realiza suma de elementos de una lista'''
    contador = 0  
    for numero in lista_numeros:
        contador = contador + numero
    return contador

lista_numeros = [1213, 1512, 511, 1, 2, 0]

resultado = suma_lista(lista_numeros)
print(resultado)
````

# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
def analyzeList (lista_numeros,N):
    is_N = N in lista_numeros
    lenght = len (lista_numeros) == N
    return is_N, lenght

lista_numeros = [1213,1512,511,1,2,0]
n = 1213
analyzeList (lista_numeros,n)
````

# Ejercicio 4

Dada una lista, cree dos funciones, una que devuelva el mayor número y otra que devuelva en menor número.



````python
def max_minN (lista_num):
    max_N = max (lista_num)
    min_N = min (lista_num)
    return max_N, min_N

lista_num = [213,152,51,34,22]

max_minN (lista_num)
````

# Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
def negative_ns (numbers_list):
    if not numbers_list: return []
    return [len([n for n in numbers_list if n <0])]

random_n = [-112,312,41,-12,523,-3,-4,0]

print (negative_ns(random_n))
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
def odd_ns (the_Nlist):
    if not the_Nlist: return []
    return [len([x for x in the_Nlist if x %2 == 0])]

random_ns = [112,-12,43,122,23,3,-4,0]

print (odd_ns(random_ns))
````

# Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
def findVowels (word):
  vowels = ['a','e','i','o','u']
  total_vowels = 0
  for v in vowels:
    if v in word:
     total_vowels += 1 
  return total_vowels

print (findVowels('sunshine')) 
````
