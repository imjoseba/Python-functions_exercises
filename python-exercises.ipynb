# Exercise 2
Write a Python function to provide the sum of the numbers in a list.

````python
def sum_list(number_list):
    '''Performs the summation of elements in a list'''
    total = 0
    for number in number_list:
        total = total + number
    return total

number_list = [1213, 1512, 511, 1, 2, 0]

result = sum_list(number_list)
print(result)
````

# Exercise 3
Given a list and a number, create a function that returns two parameters:

1) Whether that number is in the list. In such a case, it will return `TRUE`. Otherwise, it will return `False`.

2) Whether the length of the list matches that number. In such a case, it will return `TRUE`. Otherwise, it will return `False`.


````python
def check_list(number_list, target_number):
    '''Checks if a number is in the list and if the list length matches the target number'''
    is_in_list = target_number in number_list
    is_length_match = len(number_list) == target_number
    return is_in_list, is_length_match

number_list = [3, 7, 11, 15, 19]
target_number = 7

is_in_list, is_length_match = check_list(number_list, target_number)

print(f'Is {target_number} in the list? {is_in_list}')
print(f'Does the list length match {target_number}? {is_length_match}')
````

# Exercise 4
Given a list, create two functions, one that returns the largest number and another that returns the smallest number.


````python
def find_largest_number(number_list):
    '''Finds and returns the largest number in the list'''
    return max(number_list)

def find_smallest_number(number_list):
    '''Finds and returns the smallest number in the list'''
    return min(number_list)

number_list = [45, 22, 67, 89, 12, 54]

largest_number = find_largest_number(number_list)
smallest_number = find_smallest_number(number_list)

print(f'The largest number in the list is: {largest_number}')
print(f'The smallest number in the list is: {smallest_number}')
````

# Exercise 7
Create a function to count the quantity of negative numbers in a list.

````python
def count_negative_numbers(number_list):
    '''Counts the number of negative numbers in a list'''
    count = 0
    for number in number_list:
        if number < 0:
            count += 1
    return count

number_list = [45, -22, -67, 89, -12, 54]

negative_count = count_negative_numbers(number_list)

print(f'The number of negative numbers in the list is: {negative_count}')
````

# Exercise 8
Create a function to count the quantity of odd numbers in a list.

````python
def count_odd_numbers(number_list):
    '''Counts the number of odd numbers in a list'''
    count = 0
    for number in number_list:
        if number % 2 != 0:
            count += 1
    return count

number_list = [45, 22, 67, 89, 12, 54]

odd_count = count_odd_numbers(number_list)

print(f'The number of odd numbers in the list is: {odd_count}')
````

# Exercise 9
Given a word, count the number of vowels it contains.

````python
def count_vowels(word):
    '''Counts the number of vowels in a word'''
    vowels = "aeiouAEIOU"  # Define a string containing vowels
    count = 0
    for letter in word:
        if letter in vowels:
            count += 1
    return count

user_word = input("Enter a word: ")

vowel_count = count_vowels(user_word)

print(f'The word "{user_word}" contains {vowel_count} vowels.')
````

# Additional Exercise 1
Modify the password program to give the user only three attempts. If the user doesn't enter the correct password within these attempts, the program will terminate.

````python
# Define the correct password
correct_password = "secretpassword"

# Set the maximum number of attempts
max_attempts = 3

# Initialize the attempt counter
attempts = 0

while attempts < max_attempts:
    # Get the user's input
    user_password = input("Enter the password: ")

    if user_password == correct_password:
        print("Password accepted. Access granted!")
        break  # Exit the loop on successful password entry
    else:
        print("Incorrect password. Please try again.")
        attempts += 1

if attempts == max_attempts:
    print("Maximum attempts reached. Access denied.")
````

Given a matrix:

The following tasks are to be completed separately for each part:

**1**) Ask the user for a number via the console using the following command:

**2**) Ensure that the number is between 1 and 9. Display an error message if this condition is not met. The program should not continue (do not use `break` for this purpose).

**3**) Find the factors greater than the given number. To locate an element in the matrix, you can use:

*Hint*: Use nested loops, `range`, and `len`.

**4**) Return the *transposed* matrix.
[Transpose Matrix](https://upload.wikimedia.org/wikipedia/commons/e/e4/Matrix_transpose.gif). This part can be done outside the conditions from previous steps.

**Extra**) You can check if a user inputs another value (letter, symbol, or another character) using `isdigit`, `is_numeric`, or `is_decimal`. Which one would be more appropriate?

**Throughout the exercise, try to use list comprehensions whenever possible.**


````python
# Given matrix
matrix = [[1, 2], [3, 4], [5, 6], [7, 8]]

# Part 1
number = input("Please enter a number: ")

# Part 2
if not (number.isdigit() and 1 <= int(number) <= 9):
    print("Error: The number must be between 1 and 9.")
else:
    number = int(number)

    # Part 3
    factors = [element for row in matrix for element in row if element > number]

    # Part 4 (Transpose Matrix)
    transposed_matrix = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

    print(f"Factors greater than {number}: {factors}")
    print("Transposed Matrix:")
    for row in transposed_matrix:
        print(row)
````
